```python
import pandas as pd

df = pd.read_csv(
    "../data/raw/empatheticdialogues/train.csv",
    delimiter=",",  # Ensure the correct delimiter
    quoting=3,      # Disable special handling for quotation marks
    engine="python" # Use Python's parser for more flexibility
)
```

```python
data = []
currentContext = ""
speaker = 0
for index, row in df.iterrows():
    # If the context has changed, add a new conversation
    if(row['context'] != currentContext):
        speaker = 0
        currentContext = row['context']
        data.append ({
        "messages": [{"role": "system", "content": "You are an assistant that scores users' ability to prompt machine learning models to generate empathetic responses."},
                     # The first utterance in the dataset is always the prompter's
                     {"role": "user", "content": row['prompt']},
                     {"role": "user", "content": row['utterance']}]
        })
    # Otherwise, append the utterance to the last conversation
    elif(speaker == 0):
        data[len(data) - 1]["messages"].append({"role": "assistant", "content": row['utterance']})
        speaker += 1
    elif(speaker == 1):
        data[len(data) - 1]["messages"].append({"role": "user", "content": row['utterance']})
        speaker -= 1
data
```
